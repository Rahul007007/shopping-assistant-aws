```python
selected_columns = selected_columns.dropna(subset=['description'])
selected_columns.head()
```

```python
# Replace NaN values with 0 for numeric columns, and empty string for object columns
selected_columns = selected_columns.fillna({
    'discount': '$0',
})

# Convert DataFrame to list of dictionaries and add 'id' key
data = selected_columns.to_dict('records')
for idx, record in enumerate(data, start=1):
    record['id'] = idx
```

```python
from dotenv import load_dotenv
import os
load_dotenv()

api_key = os.getenv('PINECONE_API_KEY')
```

```python
from dataclasses import dataclass

@dataclass
class Config:
    api_key: str
    batch_size: int
    index_name: str = "expedite-assignment"
# Example usage
config = Config(api_key=api_key, batch_size=96)
print(config)
```

```python
# Import the Pinecone library
from pinecone.grpc import PineconeGRPC as Pinecone
from pinecone import ServerlessSpec
import time

# Initialize a Pinecone client with your API key
pc = Pinecone(api_key=config.api_key)
```

```python
if not pc.has_index(config.index_name):
    pc.create_index(
        name=config.index_name,
        dimension=1024,
        metric="cosine",
        spec=ServerlessSpec(
            cloud="aws", 
            region="us-east-1"
        ) 
    ) 

# Wait for the index to be ready
while not pc.describe_index(config.index_name).status['ready']:
    time.sleep(1)
```

```python
# batch embedding and processing
records = []
for i in range(0, len(data), config.batch_size):
    batch = data[i:i+config.batch_size]
    embeddings = pc.inference.embed(
        model="multilingual-e5-large",
        inputs=[d["description"] for d in batch],
        parameters={
            "input_type": "passage", 
            "truncate": "END"
        }
    )
    for doc, emb in zip(batch, embeddings):
        doc_copy = doc.copy()
        id = doc_copy.pop("id")
        records.append({
            "id": str(id),
            "values": emb["values"],
            "metadata": doc_copy})
    time.sleep(2)
```

```python
records[0:1][0].keys()
```

```python
index = pc.Index(config.index_name)

def chunker(seq, batch_size):
  return (seq[pos:pos + batch_size] for pos in range(0, len(seq), batch_size))

async_results = [
  index.upsert(vectors=chunk, async_req=True)
  for chunk in chunker(records, batch_size=200)
]

# Wait for and retrieve responses (in case of error)
[async_result.result() for async_result in async_results]
```

```python
print(index.describe_index_stats())
```

```python
index = pc.Index(config.index_name)
# Define your query
query = "What are the lipsticks available?"

# Convert the query into a numerical vector that Pinecone can search with
query_embedding = pc.inference.embed(
    model="multilingual-e5-large",
    inputs=[query],
    parameters={
        "input_type": "query"
    }
)

# Search the index for the three most similar vectors
results = index.query(
    # namespace="example-namespace",
    vector=query_embedding[0].values,
    top_k=3,
    include_values=False,
    include_metadata=True
)

print(results)
```

```python
results['matches'][0]
```

```python
docs_for_rerank = []

import copy
for result in results['matches']:
    # result_copy = copy.deepcopy(result)  # Copy the dictionary to avoid modifying the original
    result_metadata = result['metadata'].copy() # Extract metadata dictionary
    result_metadata['id'] = result['id']  # Add 'id' to metadata
    docs_for_rerank.append(result_metadata)  # Append only metadata

docs_for_rerank
```

```python
# Rerank the search results based on their relevance to the query
ranked_results = pc.inference.rerank(
    model="bge-reranker-v2-m3",
    query="What are the lipsticks available?",
    documents=docs_for_rerank,
    top_n=3,
    rank_fields=["description"],
    return_documents=True,
    parameters={
        "truncate": "END"
    }
)

print(ranked_results)
```

```python
# pc.delete_index(config.index_name)
for data in ranked_results.data:
    print(data.document)
```

```python
reformatted_str = ""
for data in ranked_results.data:
    result = data.document
    reformatted_str += f"Category: {result.root_category_name}\n"
    reformatted_str += f"Brand: {result.brand}\n"
    reformatted_str += f"Product ID: {result.product_id}\n"
    reformatted_str += f"Product Name: {result.product_name}\n"
    reformatted_str += f"Description: {result.description}\n"
    reformatted_str += f"Price: {result.final_price}\n"
    reformatted_str += f"Rating: {result.rating}\n"
    reformatted_str += f"Discount: {result.discount}\n"
    reformatted_str +="\n\n"

print(reformatted_str)
```
